```python
import pandas as pd
from pandasai import Agent
import os
import warnings
warnings.filterwarnings("ignore")

df = pd.read_csv("supermarket_sales.csv")
df = df[['Gender', 'Product line', 'Total']]

os.environ["PANDASAI_API_KEY"] = "Sua Chave"
```

```python
df.head()
```

```python
df.groupby('Product line')['Total'].sum()
```

```python
agent = Agent(df)
agent.chat("Which products are in Product Line")
```

```python
agent = Agent(df)
agent.chat("Calculate the total spent by each gender")
```

```python
df.groupby('Gender')['Total'].sum()
```

```python
agent = Agent(df)
agent.chat("Calculate the total spent on each product line by both the male and female gender")
```

```python
pivot_table = df.pivot_table(index='Gender',
                             columns='Product line',
                             values='Total',
                             aggfunc='sum').round(0)

pivot_table
```

```python
agent = Agent(pivot_table)
agent.chat("Make a barplot that shows how much money each gender spends on Product line Health and beauty")
```
